```python
from pyVinted import Vinted

def search_keyword(keyword):    
    vinted = Vinted("com")
    l = list()
    for i in range(10):
        l += vinted.items.search(f"https://www.vinted.com/vetements?search_text={keyword}",1000,i)
    return l
```

```python
import requests
import shutil
import os

def get_image(keyword, url, cpt) -> str:
    r = requests.get(url, stream=True)
    if r.status_code == 200:
        os.makedirs(f"vinted/{keyword}", exist_ok=True)
        if not os.path.exists(f"vinted/{keyword}/{cpt}.jpg"):
            with open(f"vinted/{keyword}/{cpt}.jpg", 'wb') as f:
                r.raw.decode_content = True
                shutil.copyfileobj(r.raw, f)
```

```python
keywords = ("robe", "pantalon", "jupe", "chaussettes", "soutien-gorge", "chemise", "veste", "manteau", "gilet", 
            "legging", "chaussures", "bottes", "talons", "baskets", "sandales", "chapeau", "bonnet", "casquette", 
            "t-shirt", "pull", "débardeur", "polo", "t-shirt manches longues", "doudoune",
)
for keyword in keywords:
    results = search_keyword(keyword)
    for result in results:
        get_image(keyword, result.photo, result.id)
```

```python
import requests
import fake_useragent
from tqdm import tqdm
```

```python
domain = "com"
VINTED_URL = f"https://www.vinted.{domain}"
VINTED_AUTH_URL = f"https://www.vinted.{domain}/auth/token_refresh"
VINTED_API_URL = f"https://www.vinted.{domain}/api/v2"
VINTED_PRODUCTS_ENDPOINT = "catalog/items"

url = f"{VINTED_API_URL}/{VINTED_PRODUCTS_ENDPOINT}"

session = requests.Session()

def get_pictures(keyword, num_pictures=10):
    user_agent = fake_useragent.UserAgent().random
    session.headers.update({"User-Agent": user_agent})
    session.post(VINTED_AUTH_URL, headers={"User-Agent": user_agent})

    num_pages = math.ceil(num_pictures / 96)
    pictures_per_page = math.ceil(num_pictures / num_pages)
    print(f"Downloading {num_pictures} pictures for {keyword} in {num_pages} pages of {pictures_per_page} items.")

    params = {"search_text": keyword, "per_page": pictures_per_page,}
    for page in tqdm(range(1, num_pages+1)):
        params['page'] = page
        response = session.get(url, params=params)
        response.raise_for_status()
        items = response.json()["items"]
        for item in items:
            try:
                yield (item["id"], item["photo"]["url"])
            except TypeError:
                continue
            
```
